### 7.3. Principal Component Analysis

**Single Value Decomposition (SVD) and Eigenvalue Decomposition (EVD)**

Checking that the diagonal matrix of eigenvalues $C_Y$ is equivalent to the square of the singular values ($\Sigma^2 = C_Y$), and the principal components can be computed from the SVD of $X$ without constructing $C_X$ (page 288).

```python
X = np.random.random((100, 3))      # Initial dataset with 100 observations made up of 3 features each
CX = np.dot(X.T, X)                 # Covariance matrix of X

U, Sdiag, VT = np.linalg.svd(X, full_matrices=False)
        # U: left-singular vectors, VT: right-singular vectors, Sdiag: diagonal of singular values matrix
CYdiag, R = np.linalg.eigh(CX)
        # CYdiag: diagonal of the sqaure of the singular values, R: principal components
```

```python
np.allclose(CYdiag, Sdiag[::-1]**2)     # Reverse Sdiag array
```

```python
VT[::-1].T / R
```

$C_Y$ and $\Sigma^2$ are equivalent; the right-hand vectors of $X$ (VT) and the eigenvectors of $C_X$ (R) are equivalent, disregarding the signs.

Using SciPy instead gives the same results:

```python
U, Sdiag, VT = sp.linalg.svd(X, full_matrices=False)
CYdiag, R = sp.linalg.eigh(CX)

np.allclose(CYdiag, Sdiag[::-1]**2)     # Reverse Sdiag array
```

```python
VT[::-1].T / R
```

**Application of Principal Component Analysis**

Performing PCA using Scikit-learn (page 289).

```python
X = np.random.normal(size=(100, 3))
R = np.random.random((3, 10))       # Projection matrix
X = np.dot(X, R)                    # Projecting X to 10D
X.shape
```

```python
pca = skl.decomposition.PCA(n_components=4)
comp = pca.fit_transform(X)             # Compute subspace projection of X to 4D
comp.shape
```

*Mean* returns the mean value for each of the 10 features, to be used to centre the data vectors before analysis.

*Components* returns the eigenvectors of the prinical components (here there are 4 with 10 features each).

*Variance* returns the eigenvalues of the principal components (again 4, and the 4th is basically 0 because X originally had 3 dimensions).

```python
mean = pca.mean_
components = pca.components_
var = pca.explained_variance_

print(f'Mean: {mean} \nComponents: {components} \nVariance: {var}')
```

### 7.4. Nonnegative Matrix Factorisation

**NMF**

Implementing NMF with Scikit-learn (page 297).

```python
X = np.random.random((100, 3))      # 100 points in 3D, all positive

nmf = skl.decomposition.NMF(n_components=3)
proj = nmf.fit_transform(X)         # Projecting to 3D
proj.shape
```

*Components* will return the components (here there are 3).

*Reconstruction error* shows how well the components capture the original data.

```python
comp = nmf.components_
err = nmf.reconstruction_err_

print(f'Components: {comp} \nError: {err}')
```

### 7.5. Manifold Learning

**Locally Linear Embedding**

Performing LLE with Scikit-learn (page 300).

```python
X = np.random.normal(size=(1000, 2))
R = np.random.random((2, 10))
X = np.dot(X, R)                # Now a 2D linear manifold in 10D space
X.shape
```

Using $LocallyLinearEmbedding$ to reconstruct the original 2D data, as in the book:

```python
lle = skl.manifold.LocallyLinearEmbedding(n_neighbors=5, n_components=2)
proj = lle.fit_transform(X[:100])
proj.shape
```

Using $locally\_linear\_embedding$ to reconstruct the 2D data, which is mostly identical to above:

```python
lle, _ = skl.manifold.locally_linear_embedding(X[:100], n_neighbors=5, n_components=2)
lle.shape
```

**IsoMap**

Implementation of the IsoMap algorithm using Scikit-learn (page 302).

```python
X = np.random.normal(size=(1000, 2))
R = np.random.random((2, 10))
X = np.dot(X, R)                # Now a 2D linear manifold in 10D space
X.shape
```

```python
iso = skl.manifold.Isomap(n_neighbors=5, n_components=2)
proj = iso.fit_transform(X)
proj.shape
```

### 7.6. Independent Component Analysis

**ICA**

Implementation of ICA with Scikit-learn (page 305).

```python
X = np.random.normal(size=(100, 2))
R = np.random.random((2, 5))
X = np.dot(X, R)            # Now 2D data in 5D space
X.shape
```

Using $FastICA$ as in the book:

```python
ica = skl.decomposition.FastICA(n_components=2)
proj = ica.fit_transform(X)
comp = ica.components_
```

**fit_transform** projects the data back into 2D space.

**Components** will return the components (here there are 2 with 5 features each).

```python
print(f'Reconstructed data: {proj.shape} \nComponents: {comp}')
```

Using $fastica$ to reconstruct data:

```python
_, _, proj = skl.decomposition.fastica(X, n_components=2)

proj.shape
```
